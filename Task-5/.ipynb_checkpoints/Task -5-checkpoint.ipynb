```python
from warnings
warnings.futurewarnings('ignore')
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv("SampleSuperstore.csv")
data.head()
```

## Exploratory Data analysis

```python
#finding the head and tail of the dataset
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data.describe()
```

```python
## finding the missing values 
print(data.isnull().sum())
```

#### There is no missing values itseems its really nice and no need for the feature engineering much 


```python
sns.pairplot(data)
```

```python
## let us find the unique values in each of these features so it would be easy for us to proceed

for val in data.columns:
    print("The column name {} has {} unique values".format(val,len(data[val].unique())))
```

```python
# let us separate the numerical and categorical features
cat =[]
num = []

for val in data.columns:
    if(len(data[val].unique()) <=20):
        cat.append(val)
    else:
        num.append(val)
print(cat,num)
```

```python
## Let us find the correlation between the features and plot it as a heatmap correlation is usually for the numerical values
sns.heatmap(data.corr(), annot=True)
```

```python
plt.figure(figsize=(20,10))
fig, ax =plt.subplots(2,4, figsize=(20,10))
sns.countplot(data['Ship Mode'], ax=ax[0][0])
sns.countplot(data['Segment'], ax=ax[0][1])
sns.countplot(data['Country'], ax=ax[0][2])
sns.countplot(data['Region'], ax=ax[0][3])
sns.countplot(data['Category'], ax=ax[1][0])
sns.countplot(data['Sub-Category'], ax=ax[1][1])
sns.countplot(data['Quantity'], ax=ax[1][2])
sns.countplot(data['Discount'], ax=ax[1][3])
fig.show()
fig.tight_layout() 
```

```python
plt.figure(figsize=(20,10))
fig, ax =plt.subplots(2,2, figsize=(20,10))
sns.relplot('City','Profit',data=data,kind="line", ax=ax[0][0])
sns.relplot('State','Profit',data=data,kind="line", ax=ax[0][1])
sns.relplot('Postal Code','Profit',data=data,kind="line", ax=ax[1][0])
sns.relplot('Sales','Profit',data=data,kind="line",ax=ax[1][1])
fig.show()
fig.tight_layout() 
```
